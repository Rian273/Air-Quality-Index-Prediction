```python
import pickle
import pandas as pd
import numpy as np
```

```python
with open("model.pkl", "rb") as file_1:
    model = pickle.load(file_1)
```

```python
model
```

```python
np.random.seed(42)

City = ['Ahmedabad', 'Amaravati', 'Amritsar', 'Bengaluru', 'Chandigarh',
       'Chennai', 'Coimbatore', 'Delhi', 'Gurugram', 'Hyderabad',
       'Jaipur', 'Kolkata', 'Lucknow', 'Patna', 'Shillong', 'Talcher',
       'Visakhapatnam']

df_inf = {
    'City' : np.random.choice(City, 50),
    'PM2.5' : np.random.uniform(0.04, 94.99, 50),
    'PM10' : np.random.uniform(0.01, 39.68, 50),
    'NO' : np.random.uniform(0.02, 36.21, 50),
    'NO2' : np.random.uniform(0.01, 32.21, 50),
    'NOx' : np.random.uniform(0.00, 47.63, 50),
    'NH3' : np.random.uniform(0.01, 32.89, 50),
    'CO' : np.random.uniform(0.00, 15.81, 50),
    'SO2' : np.random.uniform(0.01, 13.86, 50),
    'O3' : np.random.uniform(0.01, 27.73, 50),
    'Benzene' : np.random.uniform(0.00, 45.03, 50),
    'Toluene' : np.random.uniform(0.00, 44.85, 50),
    'Xylene' : np.random.uniform(0.00, 70.37, 50)
    }

df_inf = pd.DataFrame(df_inf)
```

```python
df_inf
```

```python
# Hasil prediksi
predict = model.predict(df_inf)
df_inf['AQI_predict'] = predict
df_inf
```
